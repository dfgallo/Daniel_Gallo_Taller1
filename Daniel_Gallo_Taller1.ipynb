* **Abstraer ¿qué?, ¿por qué? y ¿cómo?**

___¿Qué?___

Se busca proporcionar los datos de la cantidad de personas contagiadas, muertas y recuperadas a nivel mundial a raíz del _Covid-19._

La primera impresión que se llevará el receptor al ver la _vis_ será el número de casos positivos activos a nivel global, el número de muertes y los países con mayor cantidad de casos.

___¿Por qué?___

El receptor ante la gran desinformación y alertas de pánico provenientes de los medios tradicionales de comunicación puede acudir a esta visualización con el fin se conseguir una información precisa y sin sesgos de la situación a nivel global

En caso que el usuario requiera hacer una profundización de los datos, bastará con hacer doble click en el país de su interés o hacer click sobre las flechas inferiores de cada _chart_.

___¿Cómo?___

Los autores de la JHU para la fácil interpretación de la visualización, despliegan la información en cuadros según el ítem a interpretar (positividad, muertes, gráficas), adicional agregan un mapamundi en dónde se ve la cantidad de casos acumulados por país, logrando una fácil interpretación.

A su vez, el mapamundi deja cambiar la variable a analizar con un simple click, pasando de casos acumulados a casos activos, casos por cada 100.000 habitantes, porcentaje de mortalidad y cifras de cantidad de pruebas realizadas. De Este último inciso solo hay información existente para Estados Unidos.


* **Marcas y canales utilizados**

___Marcas___

Esta visualización emplea como marcas puntos y líneas que pueden ser evidenciadas en las gráficas de la parte inferior derecha (líneas) y en el mapamundi que despliega la información mencionada anteriormente (puntos). También hay marcas de áreas representadas en cada tabla.

___Canales___

En esta visualización sobresalen los canales de canal espacial vertical y horizontal para el atributo cuantitativo en las gráficas y tablas. Adicional, un canal de color para el atributo categórico en las diferentes estádisticas desplegadas en el mapa.

## Visualización 2  
## Every country's most popular Youtuber

[![Every country's most popular Youtuber]](C:/Users/Gallo/Desktop/DANIEL/Academic Stuff/Visualizacion python/Taller 1/youtube moneymaker.png)(https://www.accrediteddebtrelief.com/blog/wp-content/uploads/2020/09/01_Youtube-Moneymaker-Global.png)


* **Abstraer ¿qué?, ¿por qué? y ¿cómo?**

___¿Qué?___

En esta visualización, el usuario puede ver de manera exacta el Youtuber (persona) con mayor popularidad en cada país. Asímismo se proporciona la categoría y tipo de contenido que cada uno realiza. 

El propósito de la nota es dar a conocer el Youtuber que genera más ingreso en cada país y su número de seguidores, pero en la visualización principal no está ese dato con el fin de no estar sobrecargada de información la _vis_.

___¿Por qué?___

El receptor de la visuaización acudirá a esta visualización ante la incertidumbre de saber cuánto gana un Youtuber famoso, adicionalmente tendrá la capacidad de saber qué tipo de contenido es el que genera mayores ganancias a nivel global y el creador de contenido más visible en su país.

___¿Cómo?___

TopDollar, la entidad encargada de realizar la visualización despliegan la información en un mapa mundial señalizando el Youtuber más famoso de cada país, el cual está señalado con un color el cual determina la categoría del contenido que este realiza.

Con el fin de que el receptor encuentre con facilidad el país de su interés, el mapa tiene un sistema de zoom según la dirección del puntero del mouse y para los grupos de países pequeños hay apartados donde se señaliza cada uno. Adicional, en el artículo se encuentran más visualizaciones del mismo tipo pero seccionadas por contintentes y especificando los ingresos generados mensualmente por cada Youtuber.


* **Marcas y canales utilizados**

___Marcas___

En la visualización hay marcas de puntos, líneas y áreas. Las marcas de puntos son evidenciables en las convenciones de las categorías y al final de cada marca de línea que conecta el nombre del Youtuber con el país al cual pertenece. Por pultimo, las áreas son cada sección de grupos pequeños.

___Canales___

Se presentan canales de posición tanto vertical como horizontal, canales de color en la señalización de cada actegoría y canales de forma con los círculos donde se especifica la categoría y en la froma de cada país y un canal de movimiento con el zoom que se realiza al pasar el puntero.


* **Cumple o no cumple alguna de las reglas generales descritas**

La visualización cumple con todas las reglas generales propuestas por Tamara Muzner, sobresale el correcto manejo del overview y zoom propuesto por Shneiderman, aunque se le podría hacer una leve modificación presentada en el siguiente apartado.
